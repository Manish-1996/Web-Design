```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('/home/manish/Downloads/Exigent2.csv')
df.head()
```

```python
plt.figure(figsize=(50, 10))
plt.xticks(rotation=90)

x= df['Key Picked']
y1 = df['Sep25']
y2 = df['Oct2']

plt.title('Accurcy vs key picked',fontsize=18)

plt.xlabel('Accuracy', labelpad=14,fontsize=18)
plt.ylabel('Key Picked', labelpad=14,fontsize=18)


plt.plot(x,y1,label='Sep 25')
plt.plot(x,y2, label='oct 2')
plt.legend()
plt.show()
```
